```python
#### importing jovian libraries to help us with test cases ####

#### singular test case ####
from jovian.pythondsa import evaluate_test_case

#### all the test cases in an array ####
from jovian.pythondsa import evaluate_test_cases
```

```python
#### create a test dictionary with input containing the test array, query to be found, and the expected output ####
test = {
    'input': { 
        'cards': [13, 11, 10, 7, 4, 3, 1, 0], 
        'query': 7
    },
    'output': 3
}
```

```python
#### creating an empty array for tests which will contain all the edge/test cases ####
tests = []

#### appending the previously created test case to the empty array tests ####
tests.append(test)
```

```python
#### 1. Query occurs in the middle ####
tests.append({
    'input': {
        'cards': [4, 2, 1, -1],
        'query': 4
    },
    'output': 0
})
```

```python
#### 2. Query is the first element ####
tests.append({
    'input': {
        'cards': [4, 2, 1, -1],
        'query': 4
    },
    'output': 0
})
```

```python
#### 3. Query is the last element ####
tests.append({
    'input': {
        'cards': [3, -1, -9, -127],
        'query': -127
    },
    'output': 3
})
```

```python
#### 4. Cards contains just one element, i.e. query ####
tests.append({
    'input': {
        'cards': [6],
        'query': 6
    },
    'output': 0 
})
```

```python
#### 5. Cards does not contain query ####
tests.append({
    'input': {
        'cards': [9, 7, 5, 2, -9],
        'query': 4
    },
    'output': -1
})
```

```python
#### 6. Cards is empty ####
tests.append({
    'input': {
        'cards': [],
        'query': 7
    },
    'output': -1
})
```

```python
#### 7. Numbers can repeat in cards ####
tests.append({
    'input': {
        'cards': [8, 8, 6, 6, 6, 6, 6, 3, 2, 2, 2, 0, 0, 0],
        'query': 3
    },
    #### output should print the index of the position when the query appears for the first time ####
    'output': 7
})
```

```python
#### 8. Query occurs multiple times ####
tests.append({
    'input': {
        'cards': [8, 8, 6, 6, 6, 6, 6, 6, 3, 2, 2, 2, 0, 0, 0],
        'query': 6
    },
    'output': 2
})
```

```python
#### A. LINEAR SEARCH ALGORITHM ####
#### This is a brute force algorithm which doesn't take into account the time complexity of the program, for eg. if we were to find a card with minumum number of iterations in the loop(whichever you choose i.e. for, while, etc.), we can't ALWAYS use this linear search algorithm, so we need a better approach to this to optimise our program ####

def locate_card_linear_search(cards, query) :

  #### initialize index position as 0 ####
  position = 0

  print("Card : ", cards)
  print("Query : ", query)

  #### looping through the index positions of cards till position <= len(cards) - 1 ####
  for position in range(0, len(cards)) :

    print("Position : ", position)

    #### if the card element matches the query then the index position is the answer ####
    if(cards[position] == query) :
      return position

    #### increment index position by 1 if the card element at initial position doesn't match with the query ####
    position += 1

    #### if query doesn't match with any of the card elements then return -1 and break ####
    if(position == len(cards)) :
      return -1

  #### if the card length is null, i.e. empty card [] ####
  return -1
```

```python
#### evaluates the very first dictionary, i.e. test ####
evaluate_test_case(locate_card_linear_search, test)
```

```python
#### evaluates all the test cases in the array tests ####
evaluate_test_cases(locate_card_linear_search, tests)
```

```python
#### B. BINARY SEARCH ALGORITHM ####
#### Imagine you have a telephone directory with all the names sorted alphabetically, you need to find Kevin's number, you go to the middle page, tear the directory into half, you're left with names with some alphabets on either side of the torn book, now depending on whether the names with K are on the left half or the right half, you choose one of them, then go on halving the torn directories till you get the name Kevin ####

def locate_cards_binary_search(cards, query) :
  
  #### initialize low and high variables which carry the lower and upper index values of card elements respectively ####
  low = 0
  high = len(cards) - 1

  #### loop over the array cards till low <= high ####
  for low in range(0, high + 1) :

    #### find the middle index position ####
    middle = (low + high) // 2

    print("Low : ", low)
    print("High : ", high)
    print("Middle Element : ", cards[middle])

    #### if the middle element of the cards array match the query, return the middle index ####
    if(cards[middle] == query) :
      return middle 

    #### if the middle element is less than the query, which means the query lies to the left of the middle element, so we reinitialize the high value to middle - 1 and repeat the loop ####
    elif(cards[middle] < query) :
      high = middle - 1

    #### if the middle element is greater than the query, which means the query lies to the right of the middle element, so we increment the low value to middle + 1 and repeat the loop ####
    elif(cards[middle] > query) :
      low = middle + 1

  #### return -1 if card is empty [] ####
  return -1
```

```python
#### evaluates all the test cases in the array tests ####
evaluate_test_cases(locate_cards_binary_search, tests)
```
